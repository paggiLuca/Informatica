# gli indirizzi in cui ce piu di un impianto

```python
result = db.impianti.aggregate([
                                    
                                   {"$group" : {"_id": "$INDIRIZZO", "impianti" : {"$sum" : 1}}},
                                   {"$match":{"impianti":{"$gt":1}}}
              
                                   ])
df = pd.DataFrame(list(result))
df
```

# i gestori che son societa per azioni,ordinare in ordin e alfabetico

```python
result = db.impianti.aggregate([
                                  {"$match":{"gestore" : {"$regex" : "S.p.A."}}},
                                  { "$sort" : {"gestore" : 1}}
                                
                                  
                                   ])
df = pd.DataFrame(list(result))
df
```

# il numero di impianti per ogni gestore

```python
result = db.impianti.aggregate([
                                   
                                   {"$group" : {"_id": "$gestore", "nImpianti" : {"$sum" : 1}}},
                                  
                                   ])
df = pd.DataFrame(list(result))
df
```

#i gestori che hanno piu di 100 impianti.Ordinare in ordine decrescente a partire da gestore che ha piu impianti

```python
result = db.impianti.aggregate([
                                   
                                   {"$group" : {"_id": "$gestore", "nImpianti" : {"$sum" : 1}}},
                                 {"$match":{"nImpianti":{"$gt":100}}},
                                { "$sort" : {"nImpianti" : -1}},
                                   ])
df = pd.DataFrame(list(result))
df
```
