```python
# 5. calcolare il prezzo medio degli items

result = db2.vendite.aggregate([{"$unwind": "$items"},
                              {"$group": {"_id": "prezzo medio", "prezzoMedio": {"$avg":"$items.price"}}},
                              { "$project": { "prezzoMedio": { "$round": [ "$prezzoMedio", 2 ] } } }
])

df5 = pd.DataFrame(list(result))
df5
```
